```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from scipy.spatial import distance
```

```python
movies_tmp = pd.read_csv("movies_tmp.csv")
movies_tmp=movies_tmp.drop(['Unnamed: 0'], axis=1)
movies_tmp
```

```python
Y = movies_tmp[['movieId']]
X = movies_tmp.drop(['movieId','title'], axis=1)
```

```python
X['popularity'] = X['popularity'].fillna(0)
```

```python
X = X.values
Y = Y.values
```

```python
def get_weights():
    list = [0.2,3]
    list = list + [10]*19
    list = list + [2]*203
    return list
```

```python
modelKNN = KNeighborsClassifier(n_neighbors =10, weights=get_weights, algorithm='auto', p=2,metric='euclidean')
modelKNN.fit(X,Y)
```

```python
def find_similar(m_id):
    x = movies_tmp[movies_tmp.movieId ==m_id]
    x = x.drop(['movieId','title'],axis=1)
    x_list = x.values
    lm = movies_tmp.iloc[modelKNN.kneighbors(x_list,10)[1][0].tolist()]
    return lm
```

```python
def return_distance(m_id1, m_id2):
    x = movies_tmp[movies_tmp.movieId ==m_id1]
    x = x.drop(['movieId','title'],axis=1)
    x_list = x.values
    y = movies_tmp[movies_tmp.movieId ==m_id2]
    y = y.drop(['movieId','title'],axis=1)
    y_list = y.values
    return distance.euclidean(x, y)
```
