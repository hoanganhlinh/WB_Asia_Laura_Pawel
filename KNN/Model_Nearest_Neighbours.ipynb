```python
users_pr_tmp = pd.read_csv("users_preferences_new.csv")
```

```python
users_pr_tmp=users_pr_tmp.drop(['Unnamed: 0'], axis=1)
users_pr_tmp
```

```python
tp = 0
fp = 0
fn = 0
tn = 0
tp_acc = [0]*58098
fp_acc = [0]*58098
fn_acc = [0]*58098
tn_acc = [0]*58098

for i in range(1,283228):
    if i%100 == 0:
        print(i)
    pref = ast.literal_eval(users_pr_tmp[users_pr_tmp.userId == i].recommended.tolist()[0])
    notp = ast.literal_eval(users_pr_tmp[users_pr_tmp.userId == i].not_recommended.tolist()[0])
    for p in pref:
        if p <= 58097:
            near = ast.literal_eval(movies_new[movies_new.movie_index == p].nearest.tolist()[0])
            if p in near:
                near.remove(p)
            for k in near:
                if k in pref:
                    tp_acc[p] +=1
                    tp += 1
                elif k in notp:
                    fp_acc[p] +=1
                    fp += 1
    for n in notp:
        if n <= 58097:
            near = ast.literal_eval(movies_new[movies_new.movie_index == n].nearest.tolist()[0])
            if n in near:
                near.remove(n)
            for k in near:
                if k in pref:
                    fn_acc[n] +=1
                    fn += 1
                elif k in notp:
                    tn_acc[n] +=1
                    tn += 1
```

```python
acc = []
for i in range(58098):
    if (tp_acc[i]+tn_acc[i]+fp_acc[i]+fn_acc[i]) == 0:
        acc += [0]
    else:
        acc += [(tp_acc[i]+tn_acc[i])/(tp_acc[i]+tn_acc[i]+fp_acc[i]+fn_acc[i])]
```

```python
n = np.quantile(acc,0.8176)
j2 = [i for i in acc if i >= n]
len(j2)
```

### Wyniki dla całego modelu

```python
print('true positive:   ', tp)
print('false positive:  ',fp)
print('false negative:  ', fn)
print('true negative:   ',tn)
print('accuracy:        ', (tp+tn)/(tp+tn+fp+fn))
print('precision:       ', tp/(tp+fp))
print('f1:              ', 2/(1/(tp/(tp+fn))+1/(tp/(tp+fp))))
```
